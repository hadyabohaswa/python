```python
import matplotlib.pyplot as plt

# Translated information
workshops = [
    {
        "name": "Calendar Workshop and Measuring Student Progress (Port Said Hotel School)",
        "duration": "2 days",
        "total_targeted": 25,
        "attendance_rate": 88,
        "average_training_impact": 22.5,
        "average_workshop_evaluation": 78.63
    },
    {
        "name": "Planning and Designing Teaching Workshop (STA 2 - C16 School)",
        "duration": "3 days",
        "total_targeted": 16,
        "attendance_rate": 81,
        "average_training_impact": 7.6
    },
    {
        "name": "Methods and Strategies of Presentation Workshop (STA 2 - C16 School)",
        "duration": "3 days",
        "total_targeted": 21,
        "attendance_rate": 70
    }
]

# Prepare data for visualization
names = [workshop['name'] for workshop in workshops]
attendance_rates = [workshop['attendance_rate'] for workshop in workshops]
training_impacts = [workshop.get('average_training_impact', 0) for workshop in workshops]
workshop_evaluations = [workshop.get('average_workshop_evaluation', 0) for workshop in workshops]

# Function to add labels on the bars
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i] + 1, f"{y[i]}%", ha='center')

# Plotting attendance rates
plt.figure(figsize=(10, 5))
bars = plt.bar(names, attendance_rates, color='skyblue')
add_labels(names, attendance_rates)
plt.xlabel('Workshop')
plt.ylabel('Attendance Rate (%)')
plt.title('Attendance Rates of Workshops')
plt.xticks(rotation=15, ha='right')
plt.tight_layout()
plt.show()

# Plotting average training impacts
plt.figure(figsize=(10, 5))
bars = plt.bar(names, training_impacts, color='lightgreen')
add_labels(names, training_impacts)
plt.xlabel('Workshop')
plt.ylabel('Average Training Impact (%)')
plt.title('Average Training Impacts of Workshops')
plt.xticks(rotation=15, ha='right')
plt.tight_layout()
plt.show()

# Plotting average workshop evaluations
plt.figure(figsize=(10, 5))
bars = plt.bar(names, workshop_evaluations, color='lightcoral')
add_labels(names, workshop_evaluations)
plt.xlabel('Workshop')
plt.ylabel('Average Workshop Evaluation (%)')
plt.title('Average Workshop Evaluations')
plt.xticks(rotation=15, ha='right')
plt.tight_layout()
plt.show()
```
